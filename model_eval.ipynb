```python
import torch
import segmentation_models_pytorch as smp
import random, os
import numpy as np
from torch.utils.data import DataLoader, Dataset
import pandas as pd
from tqdm import tqdm
import torch.nn.functional as F

import matplotlib.pyplot as plt
```

```python
def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = True
seed_everything(42)
```

```python
import tifffile as tiff
MAX_PIXEL_VALUE = 65535 # 이미지 정규화를 위한 픽셀 최대값
# MAX_PIXEL_VALUE = 255.0 # 이미지 정규화를 위한 픽셀 최대값
def get_img_only_7bands(path, n= 6):
    img = tiff.imread(path)
    # img = img[:, :, (6, 5, 1)]#.astype(np.uint8)
    img = np.float32(img)/MAX_PIXEL_VALUE # 정규화
    # img = np.float32(img)/255 # 정규화
    # img = stretech(img[:,:,(6,5,1)])
    img = img[:,:,n] # 7채널만
    # img = new_dataset(img)
    return np.float32(img) 

def get_img_762bands2(path, flag = False):
    if flag:
        return get_img_762bands(path)
    img = tiff.imread(path)
    img = np.float32(img[:,:,(6,5,1)])  # 정규화
    sum_channels =img[:, :, 2] + img[:, :, 1]+ img[:, :, 0] + 1
    processed_img = img[:, :, 0] / sum_channels
    # processed_img = np.reshape(processed_img, (256, 256, 1))
    return np.float32(processed_img)
    
def get_img_762bands(path, flag = False):
    if flag:
        return get_img_only_7bands(path)
    img = tiff.imread(path)
    img = np.float32(img[:,:,(6,5,1)])  # 정규화
    # 채널 5와 채널 1의 값을 더함
    sum_channels = img[:, :, 2] + img[:, :, 1]  + 1
    
    # 채널 6의 값으로 나눔
    processed_img = img[:, :, 0] / sum_channels
    # 채널별 정규화 수행
    # normalized_img = (img - min_vals) / (max_vals - min_vals)
    #img = np.reshape(normalized_img, (256, 256,3))
    return np.float32(processed_img)
    
def get_mask_arr(path):
    img = tiff.imread(path)
    img = np.float32(img) # 정규화
    return img

class CustomDataset(Dataset):
    def __init__(self, imgs_path: list, masks_path: list=None, imgs_transform = None, masks_transform = None,mode = 'train'):
        self.imgs_path = imgs_path
        self.masks_path = masks_path
        self.imgs_transform = imgs_transform
        self.masks_transform = masks_transform
        self.mode = mode

    def __len__(self):
        return len(self.imgs_path)

    def __getitem__(self,idx):
        img_path = self.imgs_path[idx]
        img = get_img_762bands2(img_path)
        if self.imgs_transform:
            img = self.imgs_transform(image = img)['image']
        img = np.reshape(img, (1,256,256))
        if self.mode == 'train':
            mask_path = self.masks_path[idx]
            mask = get_mask_arr(mask_path)
            
            if self.masks_transform:
                mask = self.masks_transform(image = mask)['image']
                
            mask = np.reshape(mask, (1,256,256))
            return img, mask

        elif self.mode == 'valid': # 혹시 몰라 나눔
            mask_path = self.masks_path[idx]
            mask = get_mask_arr(mask_path)
            
            if self.masks_transform:
                mask = self.masks_transform(image = mask)['image']

            mask = np.reshape(mask, (1,256,256))
            return img, mask

        else: # test
            return img
```

```python
BASE = '/mnt/d/Jupyter-Goodyoung/forest/data'
# 데이터 위치
IMAGES_PATH = f'{BASE}/train_img/'
MASKS_PATH = f'{BASE}/train_mask/'
DEVICE = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

train_meta = pd.read_csv(f'{BASE}/train_meta.csv').sample(n=1000,random_state=42)

images_train = [os.path.join(IMAGES_PATH, image) for image in train_meta['train_img'] ]
masks_train = [os.path.join(MASKS_PATH, mask) for mask in train_meta['train_mask'] ]

train_dataset = CustomDataset(images_train, masks_train,  mode = 'train')

train_dataloader = DataLoader(
    dataset=train_dataset,
    batch_size=1,
    shuffle=True,
    num_workers=4,
    pin_memory=True
)

SAVE_PATH = f'{BASE}/train_output/'
MODEL_SAVE = f'{SAVE_PATH}/best_UNet_Pre_model.pth' # 3: efficinet 2: mobile


model = smp.Unet('timm-efficientnet-b5', encoder_weights='imagenet', classes=1, activation=None,in_channels=1)
model.load_state_dict(torch.load(MODEL_SAVE))
```

```python
def pixel_accuracy (y_true, y_pred):
    low = np.quantile(y_pred, 0.9997)
    y_pred = np.where(y_pred >  low, 1, 0) # 임계값 처리
    sum_n = np.sum(np.logical_and(y_pred, y_true))
    sum_t = np.sum(y_true)

    if (sum_t == 0):
        pixel_accuracy = 0
    else:
        pixel_accuracy = sum_n / sum_t
    return pixel_accuracy
```

```python
def calculate_iou(y_true, y_pred):
    """
    한 클래스에 대한 IoU를 계산합니다.
    """
    intersection = np.logical_and(y_true, y_pred)
    union = np.logical_or(y_true, y_pred)
    iou_score = np.sum(intersection) / np.sum(union)
    return iou_score

def calculate_miou(y_true, y_pred):
    """
    산불이 있는 경우와 없는 경우, 두 클래스에 대한 평균 IoU(mIoU)를 계산합니다.
    """
    y_pred = F.sigmoid(y_pred)
    y_pred = np.where(y_pred > 0.1, 1, 0) # 임계값 처리 -> 0.0001
    iou_fire = calculate_iou(y_true == 1, y_pred == 1)
    
    # 두 클래스에 대한 IoU 평균 계산
    return iou_fire
```

```python
model.eval()
model.to(DEVICE)
ious, miss = [], []
low = []
for imgs, masks in tqdm(train_dataloader):
    imgs = imgs.to(DEVICE)
    masks = masks.to(DEVICE)

    output = model(imgs)
    
    iou = calculate_miou(masks.cpu().detach().numpy(), output.cpu()) # mIou
    # output = output.cpu().detach().numpy()
    # masks = masks.cpu().detach().numpy()
    
    # low.append(np.quantile(output, 0.9999))
    # acc = calculate_miou(masks, output)
    if iou < 0.418:
        miss.append((iou, imgs, masks, output))
    ious.append(iou)
```
